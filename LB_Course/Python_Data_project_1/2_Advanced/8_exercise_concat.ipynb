```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data clean up
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df['job-posted_month'] = df['job_posted_date'].dt.strftime('%b')

df
```

```python
months = df['job-posted_month'].unique()
dict_months = {month: df[df['job-posted_month'] ==month] for month in months}

dict_months['Jan']
```

```python
df_q1 = pd.concat([dict_months['Jan'], dict_months['Feb'], dict_months['Mar']], ignore_index=True)

df_q1
```

```python
df_q1['job-posted_month'].value_counts().plot(kind='bar')
```

```python
df_q1.head().to_csv('quarter_1.csv')
```
